```python
grouped_data = data.groupby('release_year').size().reset_index(name='Count')
grouped_data
```

```python
grouped_data2 = data.groupby('release_year').mean()
grouped_data2
```

```python
merged_data = pd.merge(grouped_data, grouped_data2, left_on='release_year', right_on='release_year')
merged_data
```

```python
merged_data['Content spend rate'] = (merged_data['Content spend ($bn)']/merged_data['Count'])*1000000000
```

```python
merged_data
```

```python
last_6_years_spend = merged_data.iloc[3:9]['Content spend rate']
last_6_years_spend
```

```python
# H0: The content spend in 2022 will be less than the previous 6 years <= 4478411.645259134
# H1: The alternative hypothesis is the content spend in 2022 will be higher than the 6 previous years > 4478411.645259134
```

```python
from scipy import stats
from scipy.stats import ttest_1samp
import scipy.stats as st
```

```python
stat, pvalue = stats.ttest_1samp(last_6_years_spend, 2.134689e+07)
print('Stat_Score =  ',stat)
print('Pvalue =  ',pvalue/2)
```

```python
#Pvalue is really low, so we reject the equality part of our H0, and our stat_score is positive so it is higher than 0.05 so we can reject H0
```
